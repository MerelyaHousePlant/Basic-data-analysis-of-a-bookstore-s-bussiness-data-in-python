```python
top_10_vanzari = set(df.nlargest(10, 'Copii vandute')['Denumire'])

top_10_profit = set(df.nlargest(10, 'Profit generat')['Denumire'])

print(top_10_vanzari)
print(top_10_profit)
```

```python
#carti din topul celor mai bune vanzari care nu se regasesc in topul profitului
top_10_vanzari.difference(top_10_profit)
```

```python
#reuinunea dintre vanzari si profituri
top_10_vanzari.union(top_10_profit)
```

```python
#carti care se gasesc in ambele topuri
top_10_vanzari.intersection(top_10_profit)
```

```python
#cartile care nu sunt comune celor doua seturi
top_10_vanzari.symmetric_difference(top_10_profit)
```

```python
dictVanzari = df.iloc[:, [1, 2, 3, 5, 6, 7, 8]].to_dict(orient='records')
dictVanzari
```

```python
def CalculProfit(stoc, pret, copiiVandute):
  return (pret * copiiVandute) - (0.6 * pret * stoc)

for carte in dictVanzari:
    carte['Profit'] = CalculProfit(carte['Stoc'], carte['Pret'], carte['Copii vandute'])
    print(f"Cartea {carte['Denumire']} a generat un profit de {carte['Profit']}.")
```

```python
def Bibliografie(autor):
  print(f'Cărțile scrise de {autor} sunt următoarele: ')
  for carte in dictVanzari:
      if 'Autor' in carte.keys() and carte.get('Autor') == autor:
          print(carte.get('Denumire'))
  print('')
numeAutori = set([carte.get('Autor') for carte in dictVanzari])

for autor in numeAutori:
    Bibliografie(autor)
```

```python
cartiEliminate = []
for i in range(len(dictVanzari)):
    if dictVanzari[i]['Profit'] < 0:
        cartiEliminate.append(dictVanzari[i])
        print(f"Cartea {dictVanzari[i]['Denumire']} a fost eliminată pentru că avea un profit negativ.")

for carte in cartiEliminate:
    dictVanzari.pop(dictVanzari.index(carte))
```

```python
mergeInitial = pd.merge(comenzi, df, on='Id Produs', how='inner')

mergeFinal = pd.merge(mergeInitial, utilizatori, on='Id Client', how='inner')

mergeFinal.to_csv('date_comerciale.csv', index=False)

print("CSV file saved successfully.")
```

```python
print(mergeFinal.index)
print('-'*40)
print(mergeFinal.columns)
print('-'*40)
mergeFinal.head(5)
```

```python
numarComenzi = mergeFinal.groupby('nume utilizator')['Id Comanda'].count().reset_index().rename(columns={'nume utilizator': 'Nume Utilizator', 'Id Comanda': 'Numar Comenzi'})
numarComenzi
```

```python
plt.figure(figsize=(10, 6))
plt.bar(numarComenzi['Nume Utilizator'], numarComenzi['Numar Comenzi'])


plt.ylabel('Numar Comenzi')
plt.title('Numar Comenzi plasate de fiecare utilizator')

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
mergeFinal['Categorie'].replace('DIstopic', 'Distopic', inplace=True)
distributieCategorii = mergeFinal.groupby(['nume utilizator','Categorie'])['Id Comanda'].count()
distributieCategorii
```

```python
plt.rcdefaults()
plt.style.use('ggplot')
usersMultipleCategories = distributieCategorii.groupby('nume utilizator').filter(lambda x: len(x) > 1).index.get_level_values('nume utilizator').unique()

numUsers = len(usersMultipleCategories)

numCols = 3
numRows = (numUsers + numCols - 1) // numCols

fig, axes = plt.subplots(numRows, numCols, figsize=(15, 5*numRows))

fig.set_facecolor('lightgrey')

axes = np.array(axes).reshape(-1)

for i, user in enumerate(usersMultipleCategories):
    userData = distributieCategorii[user]
    ax = axes[i]

    ax.pie(userData, labels=userData.index, autopct='%1.1f%%')
    ax.set_title(f"Genuri preferate de {user}")

for j in range(i+1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

```python
evolutieVanzari = mergeFinal.groupby('data')['Pret'].sum()
evolutieVanzari.index = pd.to_datetime(evolutieVanzari.index, format='%d/%m/%Y')
evolutieVanzari = evolutieVanzari.sort_index()
evolutieVanzari
```

```python
plt.figure(figsize=(14, 6))
evolutieVanzari.index = pd.to_datetime(evolutieVanzari.index, format='%d/%m/%Y').strftime('%m-%d')
plt.plot(evolutieVanzari.index, evolutieVanzari.values, marker='o', linestyle='-')
plt.title('Trendul Venitului Zilnic')
plt.xlabel('Data')
plt.ylabel('Venitul Zilnic')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
print(evolutieVanzari.index)
```

```python
cumulativeIncome = evolutieVanzari.cumsum()


plt.figure(figsize=(14, 6))
plt.plot(cumulativeIncome.index, cumulativeIncome.values, marker='o', linestyle='-')
plt.title('Evolutia venitului din vanzari')
plt.xlabel('Data')
plt.ylabel('Venit Cumulat')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
